# Research Question

Do mismatches in sociodemographic factors measured through age gaps greater than five years, household income gaps, and differences in education levels affect the quality of committed romantic relationships in the United States for people above 18 years old?

## Background and Prior Work

There are many factors that can influence the length of a relationship. Although some may seem arbitrary, there is a general idea that certain mismatches in compatibility tend to result in shorter relationships. Sociodemographic factors such as age, economic status, race, education level, and even previous relationship histories play a role in shaping the interactions and compatibility of individuals within a relationship. 

Vogue describes a general rule of thumb that the ideal age gap should be half your age then add seven for the lower barrier, and take seven of your age then double it for the upper barrier.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Age is not always a deciding factor that makes or breaks a relationship, but the power dynamic that results can be emotionally taxing. 

A study was done by Li et al., examining the effects on a womens happiness in relation to her education compared to her husband. The data looked at a wide variety of sociodemographic variables, specifically looking at education and self esteem. The results showed a mean of -0.167 for education, referring to the fact that only 22% of women had a higher education than their counterparts.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Additionally, in comparison there is a decrease in a woman's happiness from 0.064 to 0.058 as she increases in education. The study suggests that there is a possible correlation between happiness in a relationship and the education level of both partners. The clear sign of either partner “marrying up” can release negative emotions to the relationship. Despite this, the study does not specify the effect of education on the relationship itself, which is a point to consider. 

According to The Guardian, significant income or wealth gaps in relationships can create unique challenges, often increasing power dynamics that can influence relationship stability and satisfaction. Such differences may highlight contrasting financial habits or values, potentially leading to tensions around shared expenses, lifestyle choices, or long-term goals. Studies indicate that individuals who maintain financial independence within relationships may experience less pressure tied to monetary support, but they still often face societal judgments about income disparity, especially when gender norms are challenged.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)
Previous studies and articles have mentioned many of these sociodemographic factors, but not particularly in terms of how it affects the stability or length of the relationship itself. This project seeks to explore how these factors interact to predict the length of romantic relationships, providing a quantitative analysis that aims to uncover patterns of relationship longevity across diverse sociodemographic backgrounds.

```python
# import packages
import pandas as pd
```

```python
# Loading the dataset
df = pd.read_excel('output_file.xlsx')

df.head()
```

```python
# Selecting relevant variables
columns_to_select = ['w1_ppage','w1_ppagecat','w1_ppeduc','w1_ppeducat',
                     'w1_ppethm','w1_ppgender','w1_q23','w1_ppincimp','w1_ppincimp_cat',
                     'w1_ppmarit','w1_q4','w1_q6b','w1_q9','w1_q10','w1_q34']
subset_df = df[columns_to_select]

# Renaming relevant variables for clarity
subset_df = subset_df.rename(columns={'w1_ppage':'Age',
                                      'w1_ppagecat':'Categorical Age',
                                      'w1_ppeduc':'Education',
                                      'w1_ppeducat':'Categorical Education',
                                      'w1_ppethm':'Race/Ethnicity',
                                      'w1_ppgender':'Gender',
                                      'w1_q23':'Who Earned More',
                                      'w1_ppincimp':'Total Household Income',
                                      'w1_ppincimp_cat':'Categorical Household Income',
                                      'w1_ppmarit':'Marriage Status',
                                      'w1_q4':'Partner Gender',
                                      'w1_q6b':'Partner Race',
                                      'w1_q9':'Partner Age',
                                      'w1_q10':'Partner Education',
                                      'w1_q34':'Quality of Relationship'})

subset_df.head()
```

```python
# Drop rows with missing values in the quality of relationship variable
subset_df = subset_df.dropna(subset=['Quality of Relationship'])

subset_df.head()
```

# Results

## Exploratory Data Analysis

Make sure you describe the what and why in text here as well as providing interpretation of results and context.

We decided to explore the relationships between the gaps in age, household income, and difference in education levels, and the quality of a committed romantic relationship.

```python
# Import packages
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

# Set plot parameters
plt.rcParams['figure.figsize'] = (16,10)
```

### Distributions of Variables

We want to determine the distributions of each variable of interest before establishing the necessary plots that can accurately depict the relationships we want to explore.

```python
# Exclude rows with partner's age value reported under 18
subset_df = subset_df[subset_df['Partner Age'] >= 18]
```

```python
# Distribution of Age Differences

# Create new column of calculated age differences
subset_df['Age Difference'] = abs(subset_df['Age'] - subset_df['Partner Age'])

# Visualize age differences for any potential outliers
age_gap = sns.countplot(data=subset_df, x='Age Difference', color='orange')
plt.title('Age Difference Distribution')
plt.xlabel('Age Gap Size (years)')
plt.ylabel('Number of Respondents and Their Partners')
plt.xticks(rotation=45);
```

```python
# Distribution of Total Household Incomes

# Renamed total household income values for clarity
income_order = ['Less than 5k',
                '5k-7.49k',
                '7.5k-9.99k',
                '10k-12.49k',
                '12.5k-14.99k',
                '15k-19.99k',
                '20k-24.99k',
                '25k-29.99k',
                '30k-34.99k',
                '35k-39.99k',
                '40k-49.99k',
                '50k-59.99k',
                '60k-74.99k',
                '75k-84.99k',
                '85k-99.99k',
                '100k-124.99k',
                '125k-149.99k',
                '150k-174.99k',
                '175k-199.99k',
                '200k-249.99k',
                'Greater than 250k'
               ]

subset_df['Total Household Income'] = pd.Categorical(
    subset_df['Total Household Income'],
    categories=income_order,
    ordered=True
)

# Visualize household income differences for any potential outliers
income_distribution = sns.countplot(data=subset_df, x='Total Household Income', order=income_order)

plt.title('Total Household Income Distribution')
plt.xlabel('Total Household Incomes (k)')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45);
```

```python
# Distribution of Education Level Differences

# Create new column of calculated education level differences
subset_df['Education Difference'] = abs(subset_df['Education'] - subset_df['Partner Education'])

# Visualize education level differences for any potential outliers
education_distribution = sns.countplot(data=subset_df, x='Education Difference', color='purple')
plt.title('Education Level Difference Distribution')
plt.xlabel('Education Gap Size by Difference in Levels')
plt.ylabel('Number of Respondents and Their Partners')
plt.xticks(rotation=45);
```

```python
# Distribution of Relationship Quality Scores

# Visualize relationship quality scores for any potential outliers
quality_distribution = sns.countplot(data=subset_df, x='Quality of Relationship', color='pink')
plt.title('Relationship Quality Score Distribution')
plt.xlabel('Relationship Quality Scores')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45);
```

### Relationship Quality by Age Difference

A stacked barplot was used to distinguish the varying relationship quality scores within each age difference category. The contrast from dark to light colors in the distribution demonstrate excellent to very poor quality of relationship.

```python
# Plot age difference categories vs relationship quality
df2 = subset_df.groupby(['Quality of Relationship','Age Difference'])['Quality of Relationship'].count().unstack('Age Difference').fillna(0)
sub_df2 = np.transpose(df2.div(df2.sum()))

ax = sub_df2.plot(kind='bar', stacked=True, rot=0, color=sns.color_palette("YlOrBr", as_cmap=False)[::-1])
plt.title('Relationship Quality by Age Difference')
plt.xlabel('Age Gap Size (years)')
plt.xticks(rotation=45);
```

### Relationship Quality by Household Income

A stacked barplot was used to distinguish the varying relationship quality scores within each household income category. The contrast from dark to light colors in the distribution demonstrate excellent to very poor quality of relationship.

```python
# Plot household income categories vs relationship quality
df2 = subset_df.groupby(['Quality of Relationship','Total Household Income'], observed=True)['Quality of Relationship'].count().unstack('Total Household Income').fillna(0)
sub_df2 = np.transpose(df2.div(df2.sum()))

ax = sub_df2.plot(kind='bar', stacked=True, rot=0, color=sns.color_palette("viridis", 5))
plt.title('Relationship Quality by Household Income')
plt.xlabel('Total Household Incomes (k)')
plt.xticks(rotation=45);
```

### Relationship Quality by Education Level Differences

A stacked barplot was used to distinguish the varying relationship quality scores within each education gap category. The contrast from dark to light colors in the distribution demonstrate excellent to very poor quality of relationship.

```python
# Plot household income categories vs relationship quality
df2 = subset_df.groupby(['Quality of Relationship','Education Difference'], observed=True)['Quality of Relationship'].count().unstack('Education Difference').fillna(0)
sub_df2 = np.transpose(df2.div(df2.sum()))

ax = sub_df2.plot(kind='bar', stacked=True, rot=0, color=sns.cubehelix_palette(as_cmap=False)[::-1])
plt.title('Relationship Quality by Household Income')
plt.xlabel('Total Household Incomes (k)')
plt.xticks(rotation=45);
```
